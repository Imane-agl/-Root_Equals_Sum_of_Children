Given the following list [root,left child,right child], we gonna return True if the sum of children equals to the root amd False otherwise

```python
def checkTree(root):
        if root[0] == root[1] + root[2] : return True
        else: return False
```

Example 1:

```python
root = [10,4,6]
#expected output: True
checkTree(root)
```

Example 2:

```python
root = [5,3,1]
#expected output: False
checkTree(root)
```
